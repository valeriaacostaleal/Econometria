### 2.2 Datos de precios diarios (Yahoo Finance)


```python
# Descargar precios diarios de las 11 acciones
tickers = ['NVDA', 'AAPL', 'META', 'AMD', 'INTC', 'PLTR', 'LMT', 'GD', 'JPM', 'JNJ', 'XOM']

precios = yf.download(tickers, start='2021-03-01', end='2026-03-15', auto_adjust=True)['Close']
precios = precios.dropna()

print(f"Precios diarios: {precios.shape[0]} días, {precios.shape[1]} acciones")
print(f"Periodo: {precios.index[0].date()} a {precios.index[-1].date()}")
```

```python
# Calcular retornos trimestrales reales para el panel
# Mapear cada fecha de quarter-end al precio más cercano
def get_nearest_price(prices_series, target_date, window=7):
    for offset in range(window + 1):
        for d in [target_date + pd.Timedelta(days=offset),
                  target_date - pd.Timedelta(days=offset)]:
            if d in prices_series.index:
                return prices_series.loc[d]
    return np.nan

panel['StockReturn'] = np.nan
for tk in tickers:
    if tk not in precios.columns:
        continue
    mask = panel['Ticker'] == tk
    dates = panel.loc[mask, 'Date'].values
    for i, dt in enumerate(dates):
        dt = pd.Timestamp(dt)
        price_now = get_nearest_price(precios[tk], dt)
        if i == 0 or pd.isna(price_now):
            continue
        prev_dt = pd.Timestamp(dates[i-1])
        price_prev = get_nearest_price(precios[tk], prev_dt)
        if not pd.isna(price_prev) and price_prev > 0:
            panel.loc[mask & (panel['Date'] == dt), 'StockReturn'] = (price_now / price_prev - 1) * 100

print(f"Retornos trimestrales calculados: {panel.StockReturn.notna().sum()} de {len(panel)}")
```

## 4. Modelos de Panel Data


```python
# Preparar datos para modelos de panel
# Eliminar observaciones con NAs en variables clave
vars_modelo = ['StockReturn', 'RevenueGrowthYoY', 'NetMargin', 'ROE_TTM', 'LogAssets']
panel_clean = panel.dropna(subset=vars_modelo).copy()

# Crear MultiIndex para linearmodels
panel_clean = panel_clean.set_index(['Ticker', 'Date'])

print(f"Observaciones para modelos: {len(panel_clean)}")
print(f"Empresas: {panel_clean.index.get_level_values(0).nunique()}")
print(f"Trimestres: {panel_clean.index.get_level_values(1).nunique()}")
```

### 4.1 Pooled OLS
El Pooled OLS trata todas las observaciones como independientes, sin considerar la estructura de panel. Sirve como baseline.


```python
# Definir variables
y = panel_clean['StockReturn']
X = panel_clean[['RevenueGrowthYoY', 'NetMargin', 'ROE_TTM', 'LogAssets']]
from statsmodels.tools import add_constant
X_const = add_constant(X)

# Pooled OLS
pooled = PooledOLS(y, X_const).fit(cov_type='clustered', cluster_entity=True)
print("=" * 70)
print("MODELO 1: POOLED OLS")
print("=" * 70)
print(pooled.summary)
```

### Interpretación: Pooled OLS

El modelo Pooled OLS arroja un R² de 0.144, lo que significa que los cuatro fundamentales trimestrales (crecimiento de revenue YoY, margen neto, ROE y tamaño) explican el 14.4% de la variación en el retorno trimestral de las acciones.

Las variables significativas son:

- **Net Margin (β = 0.328, p < 0.001):** Es el predictor más fuerte. Por cada punto porcentual de mejora en margen neto, el retorno trimestral sube 0.33 puntos. Esto tiene sentido porque el margen neto captura la rentabilidad real de la empresa — cuando mejora, los inversionistas ajustan sus expectativas al alza.

- **LogAssets (β = -3.417, p = 0.0007):** El signo negativo indica que las empresas más grandes tienden a tener menores retornos trimestrales. Esto es consistente con el efecto de tamaño (size effect) documentado en finanzas: las empresas más pequeñas (como PLTR con retorno promedio de 20.85%) generan mayores retornos que las mega-caps (como AAPL con 5%).

- **ROE_TTM (β = -0.069, p = 0.028):** El signo negativo es contraintuitivo a primera vista, pero se explica porque empresas con ROE muy alto (como NVDA con 76.75% o LMT con 71.74%) ya tienen esas expectativas incorporadas en el precio. Un ROE alto no garantiza retornos futuros altos — lo que importa es el *cambio* en ROE, no su nivel.

- **RevenueGrowthYoY (β = 0.027, p = 0.061):** Está al borde de la significancia. El signo positivo es intuitivo: mayor crecimiento de revenue debería traducirse en mejores retornos. Que no sea significativo al 5% sugiere que el mercado ya descuenta parte del crecimiento esperado antes de que se reporte.

El R² Between de 0.716 es alto, lo que indica que las diferencias promedio entre empresas (tech vs defensa vs energía) explican mucho de la variación. Esto es consistente con lo que vimos en las regresiones de retornos diarios, donde la correlación entre sectores era baja o negativa.


### 4.2 Efectos Fijos (Fixed Effects)
Controla por características no observables de cada empresa que no cambian en el tiempo (sector, cultura corporativa, tipo de producto). Es el modelo preferido cuando sospechamos que estas diferencias están correlacionadas con las variables independientes.


```python
# Fixed Effects
fe = PanelOLS(y, X, entity_effects=True).fit(cov_type='clustered', cluster_entity=True)
print("=" * 70)
print("MODELO 2: EFECTOS FIJOS (Fixed Effects)")
print("=" * 70)
print(fe.summary)
```

### 4.3 Efectos Aleatorios (Random Effects)
Asume que las diferencias entre empresas no están correlacionadas con las variables explicativas. Es más eficiente si el supuesto se cumple.


```python
# Random Effects
re = RandomEffects(y, X_const).fit(cov_type='clustered', cluster_entity=True)
print("=" * 70)
print("MODELO 3: EFECTOS ALEATORIOS (Random Effects)")
print("=" * 70)
print(re.summary)
```

### 4.4 Test de Hausman: FE vs RE


```python
# Test de Hausman manual
# H0: RE es consistente (diferencias no son sistemáticas)
# H1: FE es necesario (diferencias son sistemáticas)
from scipy import stats

# Extraer coeficientes comunes
common_vars = [v for v in fe.params.index if v in re.params.index]
b_fe = fe.params[common_vars].values
b_re = re.params[common_vars].values
var_diff = np.diag(fe.cov[common_vars].loc[common_vars].values - re.cov[common_vars].loc[common_vars].values)

# Hausman statistic
diff = b_fe - b_re
hausman_stat = float(diff @ np.linalg.pinv(np.diag(np.abs(var_diff))) @ diff)
hausman_pval = 1 - stats.chi2.cdf(hausman_stat, df=len(common_vars))

print("Test de Hausman (FE vs RE)")
print("=" * 40)
print(f"Chi-cuadrado: {hausman_stat:.4f}")
print(f"p-value: {hausman_pval:.4f}")
print(f"Grados de libertad: {len(common_vars)}")
if hausman_pval < 0.05:
    print("→ Se rechaza H0: usar Efectos Fijos")
    modelo_preferido = "FE"
else:
    print("→ No se rechaza H0: Efectos Aleatorios es consistente")
    modelo_preferido = "RE"
```

### Interpretación: Efectos Fijos vs Aleatorios

**Efectos Fijos (R² Within = 0.057):** Al controlar por empresa (es decir, al remover las diferencias estructurales entre compañías), el poder explicativo cae dramáticamente a 5.7%. Solo Net Margin sigue siendo significativo (β = 0.282, p = 0.002). Esto indica que *dentro de cada empresa*, las variaciones trimestrales en margen neto sí predicen el retorno, pero las demás variables no aportan información adicional una vez que se controla por la identidad de la empresa.

El F-test de poolabilidad (p = 0.232) no rechaza la hipótesis de que los efectos fijos sean iguales a cero, lo que sugiere que las diferencias entre empresas no son estadísticamente significativas en este panel — posiblemente porque con solo 11 empresas y 17 trimestres el panel es relativamente pequeño.

**Efectos Aleatorios (R² = 0.127):** Mantiene resultados similares al Pooled, con Net Margin, LogAssets y ROE como significativas.

**Test de Hausman (χ² = 7.805, p = 0.099):** No se rechaza la hipótesis nula al 5%, lo que indica que los Efectos Aleatorios son consistentes y eficientes. Esto significa que las diferencias no observables entre empresas (sector, modelo de negocio, etc.) no están correlacionadas con las variables explicativas, por lo que RE es el modelo preferido.

**¿Qué modelo usar?** El modelo de Efectos Aleatorios es el preferido por el test de Hausman. La conclusión principal es que el **margen neto es el fundamental trimestral más relevante** para explicar retornos, independientemente de la especificación. Esto conecta con el análisis de regresiones diarias: NVDA tiene el margen neto promedio más alto del portafolio (42.71%) y también el retorno trimestral promedio más alto (17.49%).

# PARTE 2: Series de Tiempo y Forecast

En esta sección modelamos cada una de las 11 acciones con modelos ARIMA para capturar la estructura temporal de los precios y generar pronósticos (forecast).

Para cada acción se aplican:
- Pruebas de estacionariedad (ADF y KPSS)
- Análisis de autocorrelación (ACF y PACF)
- Selección automática del mejor modelo ARIMA
- Forecast a 30 días con intervalos de confianza


## 5. Preparación de datos de precios


```python
# Usar precios de cierre diarios ya descargados
# Calcular log-precios y retornos
log_precios = np.log(precios)
retornos_diarios = precios.pct_change().dropna() * 100

print(f"Precios diarios: {precios.shape[0]} observaciones")
print(f"Periodo: {precios.index[0].date()} a {precios.index[-1].date()}")
print(f"\nVolatilidad diaria (desv. est. retornos %):")
vol = retornos_diarios.std().sort_values(ascending=False)
for tk, v in vol.items():
    print(f"  {tk}: {v:.2f}%")
```

## 6. Análisis de Series de Tiempo por Acción


```python
def analizar_accion(ticker, precios_df, forecast_days=30):
    """Análisis completo de series de tiempo para una acción."""
    serie = precios_df[ticker].dropna()
    retornos = serie.pct_change().dropna() * 100

    print(f"\n{'='*70}")
    print(f"ANÁLISIS DE SERIES DE TIEMPO: {ticker}")
    print(f"{'='*70}")
    print(f"Observaciones: {len(serie)} | Periodo: {serie.index[0].date()} a {serie.index[-1].date()}")

    # --- Pruebas de Estacionariedad ---
    print(f"\n--- Pruebas de Estacionariedad (precios de cierre) ---")

    # ADF
    adf_result = adfuller(serie, autolag='AIC')
    adf_stat, adf_pval = adf_result[0], adf_result[1]
    print(f"ADF Statistic: {adf_stat:.4f} | p-value: {adf_pval:.4f}", end="")
    print(f" → {'Estacionaria' if adf_pval < 0.05 else 'NO estacionaria'}")

    # KPSS
    kpss_result = kpss(serie, regression='ct', nlags='auto')
    kpss_stat, kpss_pval = kpss_result[0], kpss_result[1]
    print(f"KPSS Statistic: {kpss_stat:.4f} | p-value: {kpss_pval:.4f}", end="")
    print(f" → {'NO estacionaria' if kpss_pval < 0.05 else 'Estacionaria'}")

    # --- ACF y PACF ---
    fig, axes = plt.subplots(2, 2, figsize=(14, 8))
    fig.suptitle(f'{ticker} — Estructura Temporal', fontsize=14, fontweight='bold')

    axes[0, 0].plot(serie, color=palette[0], linewidth=0.7)
    axes[0, 0].set_title('Precio de Cierre')
    axes[0, 0].set_ylabel('USD')

    axes[0, 1].plot(retornos, color=palette[1], linewidth=0.5, alpha=0.7)
    axes[0, 1].axhline(y=0, color='gray', linestyle='--', alpha=0.5)
    axes[0, 1].set_title('Retornos Diarios (%)')

    plot_acf(serie, lags=30, ax=axes[1, 0], color=palette[2])
    axes[1, 0].set_title('ACF — Precios')

    plot_pacf(serie, lags=30, ax=axes[1, 1], color=palette[3], method='ywm')
    axes[1, 1].set_title('PACF — Precios')

    plt.tight_layout()
    plt.show()

    # --- Modelo ARIMA con auto_arima ---
    print(f"\n--- Selección de Modelo ARIMA (auto_arima) ---")
    modelo_auto = auto_arima(serie, seasonal=False, stepwise=True,
                             suppress_warnings=True, error_action='ignore',
                             max_p=5, max_q=5, max_d=2,
                             information_criterion='aic', trace=False)

    order = modelo_auto.order
    print(f"Mejor modelo: ARIMA{order}")
    print(f"AIC: {modelo_auto.aic():.2f}")
    print(f"BIC: {modelo_auto.bic():.2f}")

    # Ajustar con statsmodels para diagnósticos completos
    modelo = ARIMA(serie, order=order).fit()
    print(f"\nResumen del modelo:")
    print(modelo.summary().tables[1])

    # Diagnósticos
    residuos = modelo.resid
    lb_stat = sm_diagnostic_ljungbox(residuos, lags=[10], return_df=True)
    print(f"\nLjung-Box (lag 10): stat={lb_stat['lb_stat'].values[0]:.2f}, p-value={lb_stat['lb_pvalue'].values[0]:.4f}")

    # --- Forecast ---
    print(f"\n--- Forecast a {forecast_days} días ---")
    forecast_result = modelo.get_forecast(steps=forecast_days)
    forecast_mean = forecast_result.predicted_mean
    forecast_ci = forecast_result.conf_int(alpha=0.05)

    last_date = serie.index[-1]
    forecast_dates = pd.bdate_range(start=last_date + pd.Timedelta(days=1), periods=forecast_days)

    print(f"Último precio: ${serie.iloc[-1]:.2f}")
    print(f"Forecast día 5: ${forecast_mean.iloc[4]:.2f}")
    print(f"Forecast día 15: ${forecast_mean.iloc[14]:.2f}")
    print(f"Forecast día 30: ${forecast_mean.iloc[-1]:.2f}")

    # Gráfica de forecast
    fig, ax = plt.subplots(figsize=(14, 6))
    # Últimos 60 días + forecast
    serie_tail = serie.iloc[-60:]
    ax.plot(serie_tail.index, serie_tail.values, color=palette[0], linewidth=1.5, label='Histórico')
    ax.plot(forecast_dates, forecast_mean.values, color=palette[3], linewidth=2, label='Forecast')
    ax.fill_between(forecast_dates,
                     forecast_ci.iloc[:, 0].values,
                     forecast_ci.iloc[:, 1].values,
                     alpha=0.2, color=palette[3], label='IC 95%')
    ax.axvline(x=last_date, color='gray', linestyle='--', alpha=0.5)
    ax.set_title(f'{ticker} — Forecast ARIMA{order} ({forecast_days} días)', fontsize=13)
    ax.set_ylabel('Precio (USD)')
    ax.legend()
    plt.tight_layout()
    plt.show()

    return {
        'ticker': ticker,
        'adf_stat': adf_stat, 'adf_pval': adf_pval,
        'kpss_stat': kpss_stat, 'kpss_pval': kpss_pval,
        'order': order,
        'aic': modelo_auto.aic(),
        'forecast_30d': forecast_mean.iloc[-1],
        'last_price': serie.iloc[-1]
    }

# Importar diagnóstico de Ljung-Box
from statsmodels.stats.diagnostic import acorr_ljungbox as sm_diagnostic_ljungbox
```

```python
# Correr análisis para las 11 acciones
resultados_ts = []
for tk in tickers:
    try:
        res = analizar_accion(tk, precios)
        resultados_ts.append(res)
    except Exception as e:
        print(f"Error en {tk}: {e}")
```

## 7. Resumen de resultados — Series de Tiempo


```python
# Tabla resumen
df_resumen = pd.DataFrame(resultados_ts)
df_resumen['Estacionaria_ADF'] = df_resumen['adf_pval'].apply(lambda x: 'Sí' if x < 0.05 else 'No')
df_resumen['Estacionaria_KPSS'] = df_resumen['kpss_pval'].apply(lambda x: 'No' if x < 0.05 else 'Sí')
df_resumen['ARIMA'] = df_resumen['order'].apply(lambda x: f"({x[0]},{x[1]},{x[2]})")
df_resumen['Forecast_30d'] = df_resumen['forecast_30d'].round(2)
df_resumen['Ultimo_Precio'] = df_resumen['last_price'].round(2)
df_resumen['Cambio_%'] = ((df_resumen['forecast_30d'] / df_resumen['last_price'] - 1) * 100).round(2)

print("Resumen de Modelos de Series de Tiempo")
print("=" * 90)
cols_show = ['ticker', 'Estacionaria_ADF', 'Estacionaria_KPSS', 'ARIMA', 'aic',
             'Ultimo_Precio', 'Forecast_30d', 'Cambio_%']
print(df_resumen[cols_show].to_string(index=False))
```

### Interpretación: Resultados de Series de Tiempo

#### Estacionariedad
Las 11 acciones del portafolio son **no estacionarias** según ambas pruebas (ADF y KPSS), lo que confirma que los precios de cierre siguen tendencias y no oscilan alrededor de una media constante. Esto es esperado en series financieras y justifica el uso de modelos ARIMA con al menos una diferenciación (d ≥ 1).

#### Modelos seleccionados

Los modelos seleccionados por `auto_arima` revelan tres perfiles distintos en el portafolio:

**1. Caminata aleatoria pura — ARIMA(0,1,0):** AMD, INTC, PLTR y XOM reciben el modelo más simple posible, donde el mejor pronóstico del precio de mañana es el precio de hoy. Esto es evidencia de eficiencia de mercado en estas acciones — toda la información disponible ya está incorporada en el precio y no hay estructura temporal explotable. Para el portafolio de HowTheMarketWorks, esto implica que intentar hacer timing con estas acciones es inútil; la estrategia debe basarse en fundamentales de largo plazo, no en patrones técnicos.

**2. Estructura AR leve — ARIMA(1,1,0) o ARIMA(4,1,0):** AAPL recibe ARIMA(1,1,0) con AR(1) = 0.052 (significativo, p = 0.003) y NVDA recibe ARIMA(4,1,0) con cuatro rezagos significativos. Esto sugiere que estas acciones tienen una leve persistencia en sus movimientos — si subieron ayer, tienden ligeramente a subir hoy. Sin embargo, el efecto es muy pequeño (coeficientes < 0.1), lo que lo hace económicamente irrelevante para trading. Lo interesante es que NVDA tiene más estructura temporal que las demás, posiblemente porque los días de earnings, lanzamientos de productos y noticias de AI generan efectos que persisten por varios días.

**3. Modelos más complejos — ARIMA(2,1,2):** GD y JPM reciben modelos con componentes AR y MA, indicando que sus precios responden tanto a valores pasados como a shocks previos. GD muestra el forecast más dinámico del portafolio (de $351.52 a $353.90), lo que refleja una leve tendencia alcista captada por el modelo.

**Caso especial — JNJ con ARIMA(5,2,0):** JNJ recibe un modelo inusual con dos diferenciaciones, y el Ljung-Box (p = 0.000) indica que los residuos aún tienen autocorrelación, es decir, el modelo no captura toda la estructura temporal. Esto puede deberse a la separación de Kenvue (spin-off en 2023) que creó un quiebre estructural en la serie de precios. Para ser rigurosos, JNJ podría modelarse mejor usando solo datos post-spin-off, o con un modelo que incorpore el quiebre.

#### Forecasts
Los forecasts a 30 días son esencialmente planos para la mayoría de las acciones, lo cual es **exactamente lo que debe pasar** con modelos tipo caminata aleatoria. Si un modelo ARIMA de precios de cierre predijera con confianza que una acción va a subir 5% en 30 días, habría una oportunidad de arbitraje obvia — y en mercados líquidos como estos, eso no existe. Los intervalos de confianza se amplían con el horizonte, reflejando la incertidumbre creciente.

La excepción parcial es JNJ, cuyo forecast apunta a una ligera baja ($241.52 → $239.63, -0.78%), que refleja la tendencia negativa captada por la doble diferenciación.


## 8. Análisis de Cointegración


```python
# Prueba de cointegración por pares (Engle-Granger)
from itertools import combinations

print("Análisis de Cointegración por Pares")
print("=" * 50)

pares_coint = []
for t1, t2 in combinations(tickers, 2):
    if t1 not in precios.columns or t2 not in precios.columns:
        continue
    s1 = precios[t1].dropna()
    s2 = precios[t2].dropna()
    idx = s1.index.intersection(s2.index)
    s1, s2 = s1.loc[idx], s2.loc[idx]

    _, pval, _ = coint(s1, s2)
    pares_coint.append({'Par': f'{t1}-{t2}', 'p_value': round(pval, 4),
                        'Cointegrados': 'Sí' if pval < 0.05 else 'No'})

df_coint = pd.DataFrame(pares_coint)

# Mostrar solo los cointegrados primero
coint_si = df_coint[df_coint.Cointegrados == 'Sí'].sort_values('p_value')
print(f"\nPares cointegrados (p < 0.05): {len(coint_si)} de {len(df_coint)}")
if len(coint_si) > 0:
    print(coint_si.to_string(index=False))
else:
    print("Ningún par muestra cointegración significativa.")

print(f"\nTodos los pares:")
print(df_coint.sort_values('p_value').to_string(index=False))
```

### Interpretación: Cointegración

De los 55 pares posibles, solo **3 muestran cointegración significativa** (p < 0.05):

| Par | p-value | Interpretación |
|-----|---------|----------------|
| NVDA-JPM | 0.0174 | Relación de largo plazo entre semiconductores y el sector financiero, posiblemente porque ambos se benefician de un entorno económico favorable (más inversión en tech + más actividad financiera). |
| LMT-XOM | 0.0222 | La relación más intuitiva: defensa y energía comparten exposición a riesgos geopolíticos. Cuando hay tensiones internacionales, ambos sectores se benefician. |
| NVDA-AAPL | 0.0282 | Ambas son mega-caps tech que capturan el sentimiento general del sector tecnológico, aunque sus drivers fundamentales son diferentes (AI/data centers vs iPhone/servicios). |

La baja cantidad de pares cointegrados (3 de 55 = 5.5%) confirma que el portafolio tiene **diversificación real**: la mayoría de las acciones se mueven de forma independiente en el largo plazo. Esto es positivo para la gestión de riesgo — un portafolio donde todo estuviera cointegrado ofrecería poca protección ante caídas sectoriales.

Notablemente, NVDA-AMD **no está cointegrada** (p = 0.54), a pesar de tener la correlación diaria más alta (0.69) del portafolio. Esto ilustra una distinción importante: correlación alta no implica cointegración. Los precios de NVDA y AMD se mueven juntos en el corto plazo (correlación), pero en el largo plazo pueden divergir sin límite (sin cointegración). De hecho, NVDA ha superado a AMD por un margen enorme en 5 años gracias a su dominancia en AI, lo que explica la ausencia de equilibrio de largo plazo entre ambas.

# PARTE 3: GARCH — Modelado de Volatilidad (Puntos Extra)

Los modelos GARCH capturan la volatilidad condicional (clustering de volatilidad) que los modelos ARIMA no modelan. En series financieras, los periodos de alta volatilidad tienden a agruparse, lo cual es relevante para la gestión de riesgo del portafolio.


```python
def analizar_garch(ticker, precios_df):
    """Modelo GARCH(1,1) para una acción."""
    serie = precios_df[ticker].dropna()
    retornos = serie.pct_change().dropna() * 100

    print(f"\n{'='*70}")
    print(f"GARCH(1,1): {ticker}")
    print(f"{'='*70}")

    # Ajustar GARCH(1,1)
    modelo = arch_model(retornos, vol='Garch', p=1, q=1, mean='AR', lags=1)
    resultado = modelo.fit(disp='off')
    print(resultado.summary())

    # Forecast de volatilidad a 10 días
    forecast = resultado.forecast(horizon=10)
    vol_forecast = np.sqrt(forecast.variance.iloc[-1].values)

    print(f"\nForecast de volatilidad (desv. est. diaria %):")
    for i, v in enumerate(vol_forecast):
        print(f"  Día {i+1}: {v:.2f}%")

    # Gráfica
    fig, axes = plt.subplots(2, 1, figsize=(14, 8))

    axes[0].plot(retornos.index, retornos.values, color=palette[1], linewidth=0.4, alpha=0.7)
    axes[0].set_title(f'{ticker} — Retornos Diarios (%)')
    axes[0].axhline(y=0, color='gray', linestyle='--', alpha=0.5)

    cond_vol = resultado.conditional_volatility
    axes[1].plot(cond_vol.index, cond_vol.values, color=palette[3], linewidth=0.7)
    axes[1].set_title(f'{ticker} — Volatilidad Condicional GARCH(1,1)')
    axes[1].set_ylabel('Volatilidad (%)')

    plt.suptitle(f'{ticker} — Análisis GARCH(1,1)', fontsize=14, fontweight='bold', y=1.02)
    plt.tight_layout()
    plt.show()

    return resultado
```

```python
# Aplicar GARCH a las acciones más volátiles del portafolio
# (NVDA como variable dependiente + las más volátiles)
acciones_garch = ['NVDA', 'PLTR', 'AMD', 'META']

print("Modelos GARCH(1,1) para las acciones más volátiles del portafolio:")
print("(NVDA como dependiente + top 3 volátiles)\n")

resultados_garch = {}
for tk in acciones_garch:
    try:
        resultados_garch[tk] = analizar_garch(tk, precios)
    except Exception as e:
        print(f"Error en {tk}: {e}")
```

### Interpretación: Modelos GARCH(1,1)

Los modelos GARCH capturan el clustering de volatilidad — la tendencia de los periodos de alta volatilidad a agruparse en el tiempo. Los parámetros clave son:

| Acción | ω (omega) | α (alpha) | β (beta) | α + β | Interpretación |
|--------|-----------|-----------|----------|-------|----------------|
| NVDA | 0.466 | 0.065 | 0.894 | 0.959 | Volatilidad muy persistente — un shock de volatilidad tarda mucho en disiparse |
| PLTR | 4.894 | 0.073 | 0.643 | 0.715 | Alta volatilidad base (omega grande) pero menos persistencia |
| AMD | 2.054 | 0.059 | 0.758 | 0.818 | Persistencia intermedia con volatilidad base moderada |
| META | 1.157 | 0.161 | 0.719 | 0.880 | Mayor reactividad a shocks (alpha más alto) |

**NVDA** tiene la mayor persistencia de volatilidad (α + β = 0.959, muy cercano a 1), lo que significa que cuando hay un evento que aumenta la volatilidad (como un reporte de earnings o una restricción de exportación), el efecto dura semanas. El omega bajo (0.466) indica que en periodos normales, NVDA no es especialmente volátil — la volatilidad "explosiva" solo aparece con catalizadores.

**META** tiene el alpha más alto (0.161), lo que la hace la más reactiva a shocks individuales. Esto es coherente con su historial: la caída del 26% en un solo día en febrero 2022 (tras reportar pérdida de usuarios) y la subida del 23% en noviembre 2022 muestran cómo META amplifica los eventos idiosincráticos.

**PLTR** tiene el omega más alto (4.894), indicando que su volatilidad base es estructuralmente elevada — incluso en "días tranquilos", PLTR se mueve más que las demás. Esto se explica por su trading en base a narrativas (contratos con gobierno, expansión comercial) más que fundamentales.

**Implicaciones para el portafolio:** El forecast de volatilidad muestra que todas las acciones convergen hacia su volatilidad de largo plazo en ~10 días. Para NVDA, la volatilidad esperada es ~2.95% diario, lo que equivale a movimientos de ±$5.30 sobre el precio actual ($180). Para gestión de riesgo en HowTheMarketWorks, esto sugiere que stop-losses en NVDA deberían estar al menos 3% por debajo del precio de entrada para evitar salidas prematuras por volatilidad normal.
